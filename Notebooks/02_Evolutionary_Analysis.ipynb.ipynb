### ⭐ 3. BLAST (Basic Local Alignment Search Tool) Analysis

> The APOE protein sequence was compared against protein sequences from different vertebrate species using the **`BLASTp`** program available through NCBI.
>  BLASTp identifies proteins with similar amino acid sequences and provides information about sequence conservation across species.
> This analysis was performed to examine the **evolutionary conservation of APOE** and to compare sequence similarity among representative vertebrates.


> Why BLASTp?
> 1. Compare the human APOE protein with homologous proteins from other species
> 2. Evaluate sequence conservation during evolution
> 3. Identify closely related APOE proteins using sequence similarity

> Representative vertebrate species were selected to cover different evolutionary distances from humans.
> The dataset includes five primates (**`Human, Chimpanzee, Gorilla, Orangutan, and Rhesus macaque`**), four other mammals (**`Mouse, Rat, Dog, and Cattle`**), and one teleost fish (**`Zebrafish`**) as a distant vertebrate representative.

Understanding the BLAST Results

| Term               | Description                                                                                  |
| ------------------ | -------------------------------------------------------------------------------------------- |
| Identity (%)       | Percentage of identical amino acids between the human APOE sequence and the matched protein. |
| Query Coverage (%) | Percentage of the human APOE sequence aligned with the matched protein.                      |
| E-value            | Statistical significance of the alignment. Lower values indicate more significant matches.   |
| Protein Length     | Length of the matched protein in amino acids.                                                |



Interpretation
> 1. The BLASTp results show that APOE is highly conserved among primates. Chimpanzee, gorilla, orangutan, and rhesus macaque all exhibit sequence identities **above 93%** with complete query coverage, indicating `strong evolutionary conservation`.
> 2. Other mammals, including mouse, rat, dog, and cattle, display moderate sequence identity (**approximately 68–72%**), while maintaining high query coverage.
> 3. Zebrafish shows substantially lower sequence identity (**26.25%**), reflecting its greater `evolutionary distance from mammals`.

* Overall, these findings indicate that APOE has remained highly conserved during vertebrate evolution, particularly among primates, supporting its important biological role.
