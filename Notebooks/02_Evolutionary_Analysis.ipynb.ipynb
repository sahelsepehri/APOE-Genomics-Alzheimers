### ⭐ 3. BLAST (Basic Local Alignment Search Tool) Analysis

**Why BLASTp?**

> The BLASTP compares an amino acid query sequence against a protein sequence database
For protein database searches (BLASTP and BLASTX), the default option is the non-redundant (nr) database. This consists of the combined protein records from GenBank, the Protein Data Bank (PDB), SwissProt, PIR, and PRF.


**Search Parameters Used**:

> 1. **Program**: BLASTp
> 2. **Database**: RefSeq Protein 
> 3. **Query**: Human APOE protein sequence (Homo sapiens)
> 4. **Query RefSeq accession**: NP_000032.1
> 5. **Taxonomic restriction**: Limited to 10 species
> 6. **Scoring matrix**: BLOSUM62 (standard for protein comparisons)
> 7. **E-value threshold**: 0.05 (standard significance cutoff)

**Why These Parameters?**
> - **BLOSUM62**: Balances sensitivity and specificity for evolutionarily related proteins
> - **E-value < 0.05**: Statistically significant matches (not random)
> - **Species restriction**: Focuses on our specific evolutionary question





> - To investigate the evolutionary conservation of the APOE protein, representative vertebrate species were selected across different taxonomic groups.
> - These include primates (Homo sapiens, Pan troglodytes, Gorilla gorilla gorilla, Pongo abelii, Macaca mulatta), other mammals (Canis lupus familiaris, Bos taurus, Mus musculus, Rattus norvegicus), and a distant vertebrate outgroup (Danio rerio).


**Identity (%):**	
>- The percentage of amino acids that are identical between the query (human APOE) and the subject protein. Higher = more similar.

**Query Cover (%):**
>- The percentage of the query sequence that aligns with the subject sequence. 100% means the entire human APOE protein aligns with the subject.

**E-value (The Expectation value) :**
>- The number of matches we would expect to see by chance in a database of the same size. Lower = more significant (closer to 0 is best).

**Length:**
>- The number of amino acids in the subject protein.
