```python
import pandas as pd

# create a dictionary of blastp results
blast_data = {
    'Species': ['Human', 'Bornean orangutan', 'Western lowland gorilla', 'Chimpanzee'],
    'Scientific Name' : ['Homo sapiens', 'Pongo pygmaeus', 'Gorilla gorilla gorilla', 'Pan troglodytes'],
    'Accession': ['NP_000032.1', 'Q9GLM7.1', 'Q9GLM8.1', 'NP_001009007.1'],
    'Identity (%)': [100.00, 97.48, 97.48, 97.16],
    'Query Cover (%)': [100, 100, 100, 100],
    'E-value': [0.0, 0.0, 0.0, 0.0]
}

# Turn the disctionary into the dataframe
df_blast = pd.DataFrame(blast_data)
df_blast.style.set_caption("Description table of BLASTp Results for APOE Protein")
```

Interpretation :

- The results showed high sequence similarity among vertebrate species.
- The highest similarity was observed with primates such as chimpanzee and gorilla, indicating strong evolutionary conservation of the APOE protein.

- The results show a very high degree of conservation among primates, with over 97% sequence identity.
- This suggests that the APOE protein has remained largely unchanged throughout primate evolution.
