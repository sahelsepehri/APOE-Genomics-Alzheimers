### 1.4 Reproducing Manual Findings Programmically

 ### Gene ID 

```python
from Bio import Entrez

Entrez.email = "sahelsepehri99@gmail.com"

handle = Entrez.esearch(
    db= "gene",
    term = "APOE[Gene] AND Homo sapiens [Organism]"
)

record = Entrez.read(handle)
gene_id = record ["IdList"][0]
print("Gene ID :", gene_id)
```

---

Gene Summary

```python
handle = Entrez.esummary(
    db="gene",
    id=gene_id,
    retmode="xml"
)

summary = Entrez.read(handle)

doc = summary["DocumentSummarySet"]["DocumentSummary"][0]

print("Name:", doc["Name"])
print("Description:", doc["Description"])
print("Chromosome:", doc["Chromosome"])
print("Map Location:", doc["MapLocation"])
print("Other aliases:", doc["OtherAliases"])
```

```python
data = {
    "Gene Name": doc["Name"],
    "Description": doc["Description"],
    "Chromosome": doc["Chromosome"],
    "Map Location": doc["MapLocation"],
    "Aliases": doc["OtherAliases"]
}

df = pd.DataFrame([data])
df
```
